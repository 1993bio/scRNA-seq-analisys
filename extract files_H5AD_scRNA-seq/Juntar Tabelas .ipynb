```python
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
import anndata
```

```python
df1 = pd.read_csv("ziegler_annotations.csv") # lendo a tabela 1
```

```python
df2 = pd.read_csv("ziegler_counts.csv") # lendo a tabela 2
```

```python
# comando para juntar as duas tabelas contidas na variavel df1 e df2
df3 = pd.concat([df1, df2], axis=1 )
```

```python
df3
```

```python
# outra maneira de juntar as tabelas usando o comando merge pelo nome dos barcodes de células
df4 = pd.merge(df1, df2, how ='inner', on = 'CELL_BARCODES')
```
