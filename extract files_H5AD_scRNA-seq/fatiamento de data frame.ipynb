```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ziegler_pacientes-feminino_negativos_control_WHO_0.csv") # lendo a tabela gerada anteriormente 
```

```python
df
```

```python
# letando colunas que foram criadas na escrita do csv
del df['Unnamed: 0.3']
del df['Unnamed: 0']
del df['Unnamed: 0.1']
del df['Unnamed: 0.1.1']
df
```

```python
#selecionando somente as linhas que possuem valores cujos pacientes são negativos para sars-cov-2
df_mask=df['Cohort_Disease_WHO_Score']=='Control_WHO_8' # aqui colocar as colunas e dps as linhas que deseja separar os dados
positions = np.flatnonzero(df_mask)
filtered_df=df.iloc[positions]
filtered_df
```

```python
#escrevendo um arquivo com os dados selecionados 
filtered_df.to_csv("ziegler_pacientes-feminino_neg_ctrl_WHO_8.csv")
```
